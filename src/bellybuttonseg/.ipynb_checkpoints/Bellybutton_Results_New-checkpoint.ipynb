```python
# input dictionary of filters and array of [variables],
# returns training and test accuracy and [variables] for experiments matching the [filters]
def get_data(filters,variables,depvariables,foldername):
   
    # List parameters to filter by and store     
    outdata = {}
    for v in variables:
        outdata[v] = []
    
    for v in depvariables:
        outdata[v] = []
        
    
    filternames = []
    filtervalues = []
    for k,v in filters.items():
        filternames  += [k]
        filtervalues += [v]

    # Find relevant output folders
    matcher = 'outputs_'
    names = os.listdir(foldername)
    names = [n for n in names if n[:len(matcher)]==matcher]
    names.sort()

    
    for n in names:
        try:
            results = load_parameters(foldername+n+'/results',res_types)

            parameters = load_parameters(foldername+n+'/parameters',param_types)
            
            good_to_go = True
            for filt in filternames:
                if parameters[filt] != filters[filt]:
                    good_to_go = False
                    continue # this one is filtered out, next folder!
            
            if not good_to_go:
                continue
      
            for name in variables:
                outdata[name] += [parameters[name]]
                  
            for name in depvariables:
                outdata[name] += [results[name]]
        except:
            
            print('Failure to load '+n)
            
    for key,value in outdata.items():
        outdata[key] = np.array(value)
    
    return outdata
```

```python
from BBHP_function import results
y = results
x = ['HP_train_epochs','fraction']
filts = {'S_half':12,'scales':4,'scalefactor':2}
out = get_data(filts,x,y,filepath)
```

```python


test_train = [True, False]

y_names = ['acc','seg']

fig, axs = plt.subplots(1,len(y_names),sharex='all',sharey='all')

for idx0,y_name in enumerate(y_names):


    trainerror = 1-out['train_'+y_name]
    testerror = 1-out['test_'+y_name]

    total_epochs = out['fraction']*out['HP_train_epochs']
    uni = np.unique(out['fraction'])
    for idx,f0 in enumerate(uni):
        want = (out['fraction']==f0)

        c = (idx/len(uni),0,1-idx/len(uni))

        
        
        x = total_epochs[want]
        x2 = np.array([y for _,y in sorted(zip(x,out['fraction'][want]))])
        y1 = np.array([y for _,y in sorted(zip(x,trainerror[want]))])
        y2 = np.array([y for _,y in sorted(zip(x,testerror[want]))])
        x = np.array(sorted(x))

        for a,b in zip(x,x2):
            want2 = ((x==a)*(x2 == b)).nonzero()
            xval = np.mean(x[want2])
            yval1 = np.mean(y1[want2])
            ystd1 = np.std(y1[want2])
            yval2 = np.mean(y2[want2])
            ystd2 = np.std(y2[want2])
            if test_train[1]:
                axs[idx0].errorbar(xval,yval1,fmt='s',yerr=ystd1,color=c)
            if test_train[0]:
                axs[idx0].errorbar(xval,yval2,fmt='o',yerr=ystd2,color=c)
       # axs[idx0].loglog(x,y1,'-o',color=c)
       # axs[idx0].loglog(x,y2,'-s',color=c)

    #plt.plot([.01,10],[0.04,0.04],'k-')

    #axs[idx].set_xlabel('Number of Scales')
    axs[idx0].set_xlabel('Epochs * Data Fraction')
    axs[idx0].set_ylabel('1-'+y_name);
    axs[idx0].set_xscale('log')
    axs[idx0].set_yscale('log')

    #plt.ylim([.03,.6])

    
    axs[idx0].legend({x for k,x in enumerate({'Test','Training'}) if test_train[k]});


plt.tight_layout()
```

```python
#x = load_image(filepath+'test_images/img1_copy.JPG', binarize = False, integerize = False, RGB_to_gray = True)
prediction = load_image(filepath+'outputs_22_09_07_00_23_59/predictions/img1_copy_segmented.png', binarize = False, integerize = False, RGB_to_gray = True)


mask = load_image(filepath+'masks/img1_copy.png', binarize = False, integerize = False, RGB_to_gray = True)
AOI = load_image(filepath+'areas_of_interest/img1_copy.png', binarize = False, integerize = False, RGB_to_gray = True)
bg_val = 0;
```

```python
'''
mask = true segmented values
prediction = predicted segmented values
bg_val = value of the region we call background
'''    



# make sure disconnected components are not considered the same region
(_, mask, _, _) = cv2.connectedComponentsWithStats(mask)

# shape unimportant now, just overlap.
mask = mask.flatten()
prediction = prediction.flatten()

# find areas of each unique predicted region (store for Jaccard similarity index (J) calculation later)
pred_regs, areas_pred_regions = np.unique(prediction, return_counts=True)

# get unique values of true regions
true_regs = list(set(mask))
true_regs.remove(bg_val) # background not relevant

SEG = np.zeros(len(true_regs)) # SEG score for each true region
areas_true_regions = np.zeros(len(true_regs), dtype=int) # Area of each true region


# Cycle through true regions, find overlaps, find largest overlap and keep Jaccard (if overlap > half)
for reg_ind,reg in enumerate(true_regs):

    # isolate entire true region in predicted segmentation, store total area, find unique predicted regions within
    pred_regs_local = prediction[mask==reg] # isolate true region
    R = len(pred_regs_local) # calculate its area
    areas_true_regions[reg_ind] = R 
    unique_local_regs, local_areas = np.unique(pred_regs_local, return_counts=True) # list overlapping predicted regions

    max_overlap_idx = np.argmax(local_areas) # find predicted region with highest overlap

    # if this overlap is not background AND is more than half of the true region, calculate J
    if unique_local_regs[max_overlap_idx] != bg_val and local_areas[max_overlap_idx] > R*(0.5):
        I = local_areas[max_overlap_idx] # intersection of pred and real
        P = areas_pred_regions[np.searchsorted(pred_regs,unique_local_regs[max_overlap_idx])] # whole area of pred region
        # R = whole area of real region

        SEG[reg_ind] = I/(P+R-I)

    # if no overlap is high enough, SEG[reg_ind] = 0
```

```python
reg_ind = 10
reg = true_regs[10]

pred_regs_local = prediction[mask==reg] # isolate true region
R = len(pred_regs_local) # calculate its area
areas_true_regions[reg_ind] = R 
unique_local_regs, local_areas = np.unique(pred_regs_local, return_counts=True) # list overlapping predicted regions

max_overlap_idx = np.argmax(local_areas) # find predicted region with highest overlap

# if this overlap is not background AND is more than half of the true region, calculate J
if unique_local_regs[max_overlap_idx] != bg_val and local_areas[max_overlap_idx] > R*(0.5):
    I = local_areas[max_overlap_idx] # intersection of pred and real
    P = areas_pred_regions[np.searchsorted(pred_regs,unique_local_regs[max_overlap_idx])] # whole area of pred region
    # R = whole area of real region

    SEG[reg_ind] = I/(P+R-I)
```
